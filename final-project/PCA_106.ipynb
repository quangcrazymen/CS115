# **BASIC LIBRARY IMPORT**

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
y = df['species']
X = df.drop(columns='species')
X = X.to_numpy()
df.head(5)
```

# **Step 1: Mean subtraction**

```python
def mean(x): # np.mean(X, axis = 0)  
    return sum(x)/len(x)  

def Mean_subtraction(X):
    return X - mean(X)


X_mean_sub = Mean_subtraction(X)
X_mean_sub
```

```python
def std(x): # np.std(X, axis = 0)
    return (sum((i - mean(x))**2 for i in x)/len(x))**0.5

def Standardize_data(X):
    return Mean_subtraction(X)/std(X)

X_std = Standardize_data(X)
X_std
```

# **Step 3: Eigendecomposition of the covariance matrix**

```python
def covariance(x): 
    return np.cov(X_std.T)

cov_mat = covariance(X_std)
cov_mat
```

```python
from numpy.linalg import eig

# Eigendecomposition of covariance matrix
eig_vals, eig_vecs = eig(cov_mat)
eig_vals

# Adjusting the eigenvectors (loadings) that are largest in absolute value to be positive
max_abs_idx = np.argmax(np.abs(eig_vecs), axis=0)
signs = np.sign(eig_vecs[max_abs_idx, range(eig_vecs.shape[0])])
eig_vecs = eig_vecs*signs[np.newaxis,:]
eig_vecs = eig_vecs.T
```

```python
eig_vecs
```

```python
# Sort the most important variance to the least important variance 
explained_variances = []
for i in range(len(eig_vals)):
    explained_variances.append(eig_vals[i] / np.sum(eig_vals))
 
print(np.sum(explained_variances), '\n', explained_variances)
```

```python
# We first make a list of (eigenvalue, eigenvector) tuples
eig_pairs = [(np.abs(eig_vals[i]), eig_vecs[i,:]) for i in range(len(eig_vals))]

# Then, we sort the tuples from the highest to the lowest based on eigenvalues magnitude
eig_pairs.sort(key=lambda x: x[0], reverse=True)

# For further usage
eig_vals_sorted = np.array([x[0] for x in eig_pairs])
eig_vecs_sorted = np.array([x[1] for x in eig_pairs])
```

# **Step 4: Projection**

```python
# Select top k eigenvectors
k = 3
W = eig_vecs_sorted[:k, :] # Projection matrix

print(W.shape)
```

```python
W.T
```

```python
X_proj = X_std.dot(W.T)

print(X_proj.shape)
```
